# Cleaning

### cleaning price column

```python
rm_df['Price'] = rm_df['Price'].apply(lambda x: re.search(r'(\d{1,3}(?:,\d{3})*)', x).group(1).replace(',', '') if re.search(r'(\d{1,3}(?:,\d{3})*)', x) else None)
otm_df.Price = otm_df['Price'].apply(lambda x: re.search(r'(\d{1,3}(?:,\d{3})*)', x).group(1).replace(',', '') if re.search(r'(\d{1,3}(?:,\d{3})*)', x) else None)
zpl_df['Price'] = zpl_df['Price'].apply(lambda x: re.search(r'(\d{1,3}(?:,\d{3})*)', x).group(1).replace(',', '') if re.search(r'(\d{1,3}(?:,\d{3})*)', x) else None)
```

## Extracting property_type column for zoopla and onthemarket df

```python
def get_property_type(text):
    pattern = r'(?i)(?:bed(?:room)?\s+)([\w\s]+?)(?:\s+house|\s+to\b|\b(\w+)\sto\b|\b(\w+)\sto\s+\w+|\s+for\s+(\w+)|\s+(\bplot\b))'
    match = re.search(pattern, text)
    if match:
        return match.group(1) or match.group(2) or match.group(3) or match.group(4) or match.group(5)
    else:
        return None
```

```python
otm_df['Property_type'] = otm_df['Property_type'].apply(get_property_type)
zpl_df['Property_type'] = zpl_df['Property_type'].apply(get_property_type)
```

```python
def fix_name_type(df):
    df.drop("index", axis=1, inplace=True)
    df.rename(columns = {'Location':'Address', 'Borough' : 'Location'}, inplace = True)
    df.Price = df.Price.astype("float")
    
    return df
```

```python
rm_df = fix_name_type(rm_df)
otm_df = fix_name_type(otm_df)
zpl_df = fix_name_type(zpl_df)
```

Let's add source column

```python
rm_df["Source"] = "Rightmove"
otm_df["Source"] = "Onthemarket"
zpl_df["Source"] = "Onthemarket"
```

```python
def unify_property_type(property_type):
    
    if property_type is None:
        return None
    
    words = property_type.lower().split()
    if 'end' in words and 'terrace' in words:
        return 'End Terraced'
    elif 'plot' in words or 'land' in words:
        return 'Land'
    else:
        return property_type.capitalize()
```

```python
rm_df['Property_type'] = rm_df['Property_type'].apply(unify_property_type)
otm_df['Property_type'] = otm_df['Property_type'].apply(unify_property_type)
zpl_df['Property_type'] = otm_df['Property_type'].apply(unify_property_type)
```
